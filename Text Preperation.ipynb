```python
from typing import Text
import nltk
import re
import pandas as pd
Text= pd.read_csv (r'/Users/ali/Desktop/Final Project/Raw-text.csv',encoding='latin-1')
#Text.to_csv (r'/Users/ali/Desktop/My text.csv', index=None)
ListWords=[]
Sentences= []
TaggingDF = pd.DataFrame(columns=["Words"])
for i in range(0 , len(Text)):
    ListWords= nltk.word_tokenize((re.sub(r'[.,()]', '', Text["Text"][i])).lower())
    TaggingDF = TaggingDF.append(pd.DataFrame({"Words": ListWords}))
    for j in range(0 , len(ListWords)):
        Sentences.append(("Sentence: " +"%d" %(i+1)))
        j+=1
    i+=1
TaggingDF["Sentences"] = Sentences
TaggingDF.to_csv('/Users/ali/Desktop/text12.csv')
```
